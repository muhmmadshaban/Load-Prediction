```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr=LogisticRegression(C=0.1, penalty='l2', solver='lbfgs', max_iter=1000)
lr.fit(x_train,y_train)
```

```python
lr.score(x_test,y_test)*100
```

```python
lr.score(x_train,y_train)*100
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
r=RandomForestClassifier(n_estimators=100, max_depth=5, min_samples_leaf=10,random_state=42)
r.fit(x_train,y_train)
```

```python
r.score(x_test,y_test)*100
```

```python
r.score(x_train,y_train)*100
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
d=DecisionTreeClassifier(max_depth=5, min_samples_leaf=10,random_state=42)
d.fit(x_train,y_train)
```

```python
d.score(x_test,y_test)*100
```

```python
d.score(x_train,y_train)*100
```

```python
from sklearn.svm import SVC
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
k=KNeighborsClassifier(n_neighbors=10)
k.fit(x_train,y_train)
```

```python
from sklearn.ensemble import VotingClassifier
```

```python
model=VotingClassifier(estimators=[('rf',r),('dt',d),('knn',k)],voting='soft')
model.fit(x_train,y_train)
```

```python
from sklearn.model_selection import cross_val_score
for name, model in [('Logistic Regression', lr), ('Random Forest', r), ('Decision Tree', d), ('KNN', k)]:
    scores = cross_val_score(model, x, y, cv=5)
    print(f"{name} average CV accuracy: {scores.mean() * 100:.2f}%")
```

```python
model2 = VotingClassifier(estimators=[('lr', lr),('dt', d),('knn', KNeighborsClassifier(n_neighbors=10)),('rf',r)],voting='soft')
scores = cross_val_score(model2, x,y,cv=5)
print("Ensemble CV Accuracy:", scores.mean() * 100)
```

```python
from sklearn.metrics import confusion_matrix, classification_report
model2.fit(x_train, y_train)
y_pred = model2.predict(x_test)
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
from sklearn.linear_model import LogisticRegressionCV
# L2 regularized logistic regression with cross-validation
lr_l2 = LogisticRegressionCV(
    Cs=10,
    cv=5,
    penalty='l2',
    solver='lbfgs',  # lbfgs supports only L2
    max_iter=1000,
    scoring='accuracy'
)
lr_l2.fit(x,y)
print("L2 Regularized Logistic Regression CV Accuracy: {:.2f}%".format(lr_l2.score(x, y) * 100))
```

```python
# L1 regularized logistic regression with cross-validation
lr_l1 = LogisticRegressionCV(
    Cs=10,
    cv=5,
    penalty='l1',
    solver='liblinear',  # liblinear supports L1
    max_iter=1000,
    scoring='accuracy'
)
lr_l1.fit(x,y)
print("L1 Regularized Logistic Regression CV Accuracy: {:.2f}%".format(lr_l1.score(x, y) * 100))
```

```python
from sklearn.metrics import roc_curve, auc
import matplotlib.pyplot as plt

# Generate ROC curve values
fpr, tpr, thresholds = roc_curve(y_test, y_pred)  # y_pred should be probabilities, not labels

# Calculate AUC (Area Under Curve)
roc_auc = auc(fpr, tpr)

# Plot ROC Curve
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='blue', lw=2, label=f"ROC Curve (AUC = {roc_auc:.2f})")
plt.plot([0, 1], [0, 1], color='gray', lw=1, linestyle='--', label='Random Guess')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend(loc='lower right')
plt.grid()
plt.show()
```
